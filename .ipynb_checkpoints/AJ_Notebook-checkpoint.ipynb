```python
batch_size=32
```

```python
train_data_gen = ImageDataGenerator(rescale=1/255., shear_range=0.2, zoom_range=0.2, horizontal_flip=True)
valid_data_gen = ImageDataGenerator(rescale=1/255.)
```

```python
train_generator = train_data_gen.flow_from_directory(
        '/home/ec2-user/data/dogscats/train/',
        target_size=(150, 150),
        batch_size=batch_size,
        class_mode='categorical')

validation_generator = valid_data_gen.flow_from_directory(
        '/home/ec2-user/data/dogscats/valid/',
        target_size=(150, 150),
        batch_size=batch_size,
        class_mode='categorical')
```

```python
model = Sequential()
model.add(Conv2D(32, (3,3), input_shape=(150, 150, 3), padding='same', activation='relu'))
model.add(MaxPool2D(pool_size=(2,2)))

model.add(Conv2D(32, (3,3), input_shape=(150, 150, 3), padding='same', activation='relu'))
model.add(MaxPool2D(pool_size=(2,2)))

model.add(Conv2D(64, (3,3), input_shape=(150, 150, 3), padding='same', activation='relu'))
model.add(MaxPool2D(pool_size=(2,2)))

model.add(Dropout(0.25))
model.add(Flatten())
model.add(Dense(64, activation='relu'))
model.add(Dropout(0.5))
model.add(Dense(2, activation='softmax'))
```

```python
epochs  = 50
learning_rate = 0.01
decay = learning_rate / epochs
stochastic_grad_descent = SGD(lr=learning_rate, momentum=0.9, decay=decay, nesterov=False)
model.compile(loss='binary_crossentropy', optimizer=stochastic_grad_descent, metrics=['accuracy'])
```
